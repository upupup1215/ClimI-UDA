```python
import numpy
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from tslearn.barycenters import \
    euclidean_barycenter, \
    dtw_barycenter_averaging, \
    dtw_barycenter_averaging_subgradient, \
    softdtw_barycenter
from tslearn.datasets import CachedDatasets
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Step 1 Calculation of the barycenter for different climate variables

```python
year = '2017'
day = '4'
sa = 'target'
grid_path = r'E:\jupyter\ClimI-UDA\ClimVs\{}day_{}\{}'.format(day,sa,year)
lsgrid=os.listdir(grid_path)
```

```python
CVtslistall = []
for i in range(len(lsgrid)):
    CVs_path=grid_path+'/'+str(lsgrid[i])
    lsCVs=os.listdir(CVs_path)
    CV_path = CVs_path+'/'+str(lsCVs[1]) #[0,1,2,3,4,5]
    CVts = pd.read_csv(CV_path,header=None)
    CVts = np.array(CVts)
    CVtslist = CVts.tolist()
    CVtslistall.append(CVtslist)
    
CVtsall = np.array(CVtslistall)
# print(CVtsall.shape)    
X = np.vstack((CVtsall,))

index = lsCVs[1].split('_')[0] #CVname

def plot_helper(barycenter):
    # plot all points of the data set
    for series in X:
        plt.plot(series.ravel(), "k-", alpha=.2)
    # plotbarycenter.ravel() the given barycenter of them
    A =list(barycenter.ravel())
#     print(A)
    plt.plot(barycenter.ravel(), "r-", linewidth=2)

    return list(barycenter.ravel())

# plt.grid(True) 
# plot the four variants with the same number of iterations and a tolerance of
# 1e-3 where applicable
plt.figure(figsize=(16,8),dpi=300)
plt.grid(True) 
ax1 = plt.subplot(4, 1, 1)
plt.title("Euclidean barycenter")
x1 = plot_helper(euclidean_barycenter(X))
NWdifference1 = pd.DataFrame(np.array(x1))
isExists = os.path.exists(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm1'.format(day,sa))
if not isExists:
    os.makedirs(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm1'.format(day,sa))
NWdifference1.to_csv(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm1\{}day_1_{}.csv'.format(day,sa,day,index),header=None,index = None)

plt.subplot(4, 1, 2, sharex=ax1)
plt.title("DBA (vectorized version of Petitjean's EM)")
x2 =plot_helper(dtw_barycenter_averaging(X, max_iter=1, tol=1e-3))
NWdifference2 = pd.DataFrame(np.array(x2))
isExists = os.path.exists(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm2'.format(day,sa))
if not isExists:
    os.makedirs(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm2'.format(day,sa))
NWdifference2.to_csv(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm2\{}day_2_{}.csv'.format(day,sa,day,index),header=None,index = None)

plt.subplot(4, 1, 3, sharex=ax1)
plt.title("DBA (subgradient descent approach)")
x3 =plot_helper(dtw_barycenter_averaging_subgradient(X, max_iter=50, tol=1e-3))
NWdifference3 = pd.DataFrame(np.array(x3))
isExists = os.path.exists(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm3'.format(day,sa))
if not isExists:
    os.makedirs(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm3'.format(day,sa))
NWdifference3.to_csv(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm3\{}day_3_{}.csv'.format(day,sa,day,index),header=None,index = None)

plt.subplot(4, 1, 4, sharex=ax1)
plt.title("Soft-DTW barycenter ($\gamma$=1.0)")
x4 =plot_helper(softdtw_barycenter(X, gamma=1., max_iter=50, tol=1e-3))
NWdifference4 = pd.DataFrame(np.array(x4))
isExists = os.path.exists(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm4'.format(day,sa))
if not isExists:
    os.makedirs(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm4'.format(day,sa))
NWdifference4.to_csv(r'E:\jupyter\ClimI-UDA\CVb\{}day{}_cm4\{}day_4_{}.csv'.format(day,sa,day,index),header=None,index = None)
```

# Step2 Calculation of the Source ClimI

```python
CVb_path = r'E:\jupyter\ClimI-UDA\CVb\{}daysource_cm2'.format(day,sa)
lsCVb=os.listdir(CVb_path)
```

```python
####
CVbs_path = CVb_path+'/'+str(lsCVb[0]) #[0,1,2,3,4,5]
CVbts0 = pd.read_csv(CVbs_path,header=None)
CVbts0_max = CVbts0.max(axis=0)
CVbts0_min = CVbts0.min(axis=0)
CVbts0Normlization = (CVbts0-CVbts0_min)/(CVbts0_max-CVbts0_min)
####
CVbs_path = CVb_path+'/'+str(lsCVb[1]) #[0,1,2,3,4,5]
CVbts1 = pd.read_csv(CVbs_path,header=None)
CVbts1_max = CVbts1.max(axis=0)
CVbts1_min = CVbts1.min(axis=0)
CVbts1Normlization = (CVbts1-CVbts1_min)/(CVbts1_max-CVbts1_min)
####
CVbs_path = CVb_path+'/'+str(lsCVb[2]) #[0,1,2,3,4,5]
CVbts2 = pd.read_csv(CVbs_path,header=None)
CVbts2_max = CVbts2.max(axis=0)
CVbts2_min = CVbts2.min(axis=0)
CVbts2Normlization = (CVbts2-CVbts2_min)/(CVbts2_max-CVbts2_min)
####
CVbs_path = CVb_path+'/'+str(lsCVb[3]) #[0,1,2,3,4,5]
CVbts3 = pd.read_csv(CVbs_path,header=None)
CVbts3_max = CVbts3.max(axis=0)
CVbts3_min = CVbts3.min(axis=0)
CVbts3Normlization = (CVbts3-CVbts3_min)/(CVbts3_max-CVbts3_min)
####
CVbs_path = CVb_path+'/'+str(lsCVb[4]) #[0,1,2,3,4,5]
CVbts4 = pd.read_csv(CVbs_path,header=None)
CVbts4_max = CVbts4.max(axis=0)
CVbts4_min = CVbts4.min(axis=0)
CVbts4Normlization = (CVbts4-CVbts4_min)/(CVbts4_max-CVbts4_min)
####
CVbs_path = CVb_path+'/'+str(lsCVb[5]) #[0,1,2,3,4,5]
CVbts5 = pd.read_csv(CVbs_path,header=None)
CVbts5_max = CVbts5.max(axis=0)
CVbts5_min = CVbts5.min(axis=0)
CVbts5Normlization = (CVbts5-CVbts5_min)/(CVbts5_max-CVbts5_min)
```

```python
AllCVb = pd.concat([CVbts0Normlization, CVbts1Normlization, CVbts2Normlization,CVbts3Normlization, CVbts4Normlization, CVbts5Normlization], axis=1)
AllCVb_Normlization_arr = AllCVb.values
AllCVb_Normlization_arr = AllCVb_Normlization_arr.transpose(1,0)
AllCVb_Normlization_arr = AllCVb_Normlization_arr.reshape(6,46,1) #cv number,cvts length,1
```
